```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df = pd.read_csv("employee_sales_data.csv")
X = df.drop(columns=['SalesPerformance']).values
y = df['SalesPerformance'].values
y.dtype
```

```python
plt.scatter(df['Experience'], y)
plt.xlabel('Experience')
plt.ylabel('SalesPerformance')
plt.title("Scatter plot of Experience vs SalesPerformance")
plt.plot()
```

```python
# Group by 'EducationLevel' and calculate the mean of 'SalesPerformance'
mean_sales_performance = df.groupby('EducationLevel')['SalesPerformance'].mean()

# Plot the results
mean_sales_performance.plot(kind='bar')
plt.xlabel('EducationLevel')
plt.ylabel('Average Sales Performance')
plt.title('Average Sales Performance by Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
X = df.drop(columns=['SalesPerformance'])
y = df['SalesPerformance']
for column in X.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(X[column], y)
    plt.xlabel(column)
    plt.ylabel('SalesPerformance')
    plt.title(f"Sales Performance vs {column}")
    plt.show()
```
